```python
import pickle
```

```python
import pandas as pd
```

```python
walnut_model = pickle.load(open('walnut_lrcv.pkl', 'rb'))
```

```python
types_df = pd.read_csv('types.csv.bz2', compression='bz2')
```

```python
types_df.head()
```

```python
types_df['distance_ft'] = 50
```

```python
types_df.head()
```

```python
species_df = types_df[['scientific_name', 'distance_ft']]
```

```python
species_df.head()
```

```python
species_df = species_df.dropna()
```

```python
species_df = species_df.rename(columns={'scientific_name':'s_name'})
```

```python
walnut_df = species_df
```

```python
walnut_df['survival'] = walnut_model.predict_proba(walnut_df[['s_name', 'distance_ft']])
walnut_df.sort_values('survival', ascending=False)
```

```python
walnut_df['survival'] = walnut_model.predict(walnut_df[['s_name', 'distance_ft']])

walnut_df.sort_values('survival', ascending=False)
```
